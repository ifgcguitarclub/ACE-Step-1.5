## Local Inference on GPU

```python
# @title üß† URC Brain Injection (4B Mode - Ultra Efficient)
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

# Arsitektur Kuantisasi NF4: Rahasia agar 4B terasa seperti 0.5B dalam hal memori
urc_bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

model_id = "ACE-Step/acestep-5Hz-lm-4B"

print(f"üöÄ Memasang Otak Maestro 4B ke Sasis T4...")
print("‚ö†Ô∏è Menghindari OOM Error dengan Teknik Injeksi 4-bit...")

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=urc_bnb_config,
    device_map="auto",
    low_cpu_mem_usage=True,
    trust_remote_code=True
)

print("\n" + "X"*50)
print("üî• STATUS: OTAK 4B BERHASIL AKTIF!")
print("üìä Pemakaian VRAM: Terpangkas dari 16GB menjadi ~2.9GB!")
print("‚úÖ Siap menerima perintah Sabda tingkat tinggi.")
print("X"*50)
```

```python
# @title üìù URC Sabda Generator (Lyrics & Music Structure)
prompt = "Buat lagu Rock Alternatif tentang kebangkitan mesin AI di GPU T4, lirik bahasa Indonesia, tempo cepat, emosional." # @param {type:"string"}
max_tokens = 512 # @param [256, 512, 1024]

print("üß† Otak 4B sedang merumuskan Sabda...")

inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )

hasil_sabda = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("\n" + "="*50)
print("üìú HASIL RANCANGAN MUSIK (THE SABDA):")
print("="*50)
print(hasil_sabda)
print("="*50)
print("‚úÖ Blueprint siap dikirim ke Mesin Suara (DiT)!")
```
